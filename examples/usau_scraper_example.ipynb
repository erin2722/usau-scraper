```python
# import all modules from usau-scraper
from usau_scraper import *

# import json for pretty printing
import json
```

## Team Info

```python
# You can search for one specific team using schoolName and teamName
print(json.dumps(getTeamInfo(schoolName='Columbia', teamName='Curbside'), indent = 4))
```

```python
# Or search with even more specific information for a team
print(json.dumps(getTeamInfo(
    schoolName = 'Columbia', 
    teamName = 'Baewatch', 
    genderDivision='Women', 
    state='NY', 
    competitionLevel='College', 
    competitionDivision='Division 1', 
    teamDesignation='B'), indent=4))
```

```python
# Search for the first 20 College Womens Teams in New York
print(json.dumps(getTeamInfo(
    genderDivision='Women', 
    state='NY', 
    competitionLevel='College'), indent=4))
```

## Team Schedule

```python
print(json.dumps(getTeamSchedule(schoolName='Columbia', teamName='Curbside'), indent = 4))
```

## Team Roster

```python
# Search for a team's roster for the current season
print(json.dumps(getTeamRoster(schoolName='Columbia', teamName='Curbside'), indent = 4))
```

## Team Rankings

```python
# Search for the top 20 womens college teams
print(json.dumps(getCollegeRankings(genderDivision="Women"), indent=4))
```

```python
# Search for the top 20 mixed club teams
print(json.dumps(getClubRankings(genderDivision="Mixed"), indent=4))
```

## Tournament Pool Play Results

```python
# Get the pool play results for a tournament
print(json.dumps(getTournamentPoolPlayResults("College", "Women", eventName="No Sleep Till Brooklyn", season=2023), indent=4))
```

## Tournament Bracket Play Results

```python
# Get the bracket play results for a tournament
print(json.dumps(getTournamentBracketResults("College", "Men", eventName="Centex", season=2022), indent=4))
```

## Get the Tournament Winner

```python
print(json.dumps(getTournamentWinner("College", "Women", eventName="No Sleep Till Brooklyn", state='NY', season=2022), indent=4))
```

## Advanced Example: Get the schedule for each top 20 team

```python
teams = getCollegeRankings(genderDivision="Women")["teams"]

for team in teams:
  print(json.dumps(getTeamSchedule(schoolName=team["Team"], genderDivision='Women', competitionLevel='College')["teams"][0], indent=4))
```
