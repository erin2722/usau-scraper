```python
# import json for pretty printing
import json
```

## Team Info

```python
from usau_scraper import getTeamInfo

# You can search for one specific team using schoolName and teamName
print(json.dumps(getTeamInfo(schoolName='Columbia', teamName='Curbside'), indent = 4))

# Or search with even more specific information for a team (encodings will be made easier soon)
print(json.dumps(getTeamInfo(
    schoolName = 'Columbia', 
    teamName = 'Baewatch', 
    genderDivision='Women', 
    state='NY', 
    competitionLevel='College', 
    competitionDivision='Division 1', 
    teamDesignation='B'), indent=4))

# Search for the first 20 College Womens Teams in New York
print(json.dumps(getTeamInfo(
    genderDivision='Women', 
    state='NY', 
    competitionLevel='College'), indent=4))
```

## Team Schedule

```python
from usau_scraper import getTeamSchedule

# Search for a team's schedule for the current season
print(json.dumps(getTeamSchedule(schoolName='Columbia', teamName='Curbside'), indent = 4))
```

## Team Roster

```python
from usau_scraper import getTeamRoster

# Search for a team's roster for the current season
print(json.dumps(getTeamRoster(schoolName='Columbia', teamName='Curbside'), indent = 4))
```
